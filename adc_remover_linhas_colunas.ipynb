```python
# carregando uma planilha
from openpyxl import load_workbook

workbook = load_workbook(filename='controle_da_empresa.xlsx')

# verificando as abas
sheet = workbook.active
```

```python
def print_rows():
  for row in sheet.iter_rows(values_only = True):
    print(row)
```

```python
print_rows()
```

```python
#adicionando uma coluna

sheet.insert_cols(idx=1)
print_rows()
```

```python
sheet.insert_cols(idx=3,amount=3)
print_rows()
```

```python
#deletando colunas
sheet.delete_cols(idx=3,amount=3)
print_rows()
```

```python
#deletando colunas
sheet.delete_cols(idx=1)
print_rows()
```

```python
##adicionado uma LINHA

sheet.insert_rows(idx=1)
print_rows()
```

```python
#adicionando 5 linhas de uma vez

sheet.insert_rows(idx=1, amount=5)
print_rows()
```
