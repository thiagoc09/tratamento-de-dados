```python
produtos = {}

for row in df.iter_rows(min_row=2,
                           min_col=4,
                           max_col=7,
                           values_only=True):
  product_id = row[0]
  produto = {
      "parent" : row[1],
      "title" : row[2],
      "category" : row[3]

  }
  produtos[product_id] = produto
```
