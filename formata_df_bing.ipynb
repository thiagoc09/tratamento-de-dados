```python
import pandas as pd 
df = pd.read_excel("/content/GOL_BING_DATALAKE_RELATORIO_PADRAO.xlsx", skiprows=6)
```

```python
df.head()
```

```python
def formata_bing_df(df):
      df = df.drop(columns=['Ad ID'])      
      correcao = {"Date":"Unidade de tempo","Campaign name":"Nome da campanha","Ad type":"Tipo do anúncio","Ad group":"Grupo de anúncios","Spend":"Gastos","Device type":"Tipos de dispositivo","Impressions":"Visualizações", "Clicks":"Cliques","Final URL":"URL Final"}
        
      df = df.rename(columns=correcao)
      return df
```

```python
formata_bing_df(df)
```
